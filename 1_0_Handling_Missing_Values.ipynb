```python
## Missing Values
## Missing Values occurs in a data set when some of the information is not stored  for a variable .. there are 3 mechanisms
## 1 . Missing Completely at Random (MCAR)
## 2 . Missing at Random (MAR)
## 3 . Missing Not at Random (MNAR)
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
##Checking missing Values
df.isnull()
```

```python
df.isnull().sum()
```

```python
## Delete the rows or data point to handle missingt Values

df.shape
```

```python
df.dropna().shape
```

```python
## column wise
## Deletion

df.dropna(axis=1)
```

```python
df.dropna(axis=1).shape
```

```python
### Imputation Missing  Values
```

```python
## 1 - Mean Value Imputation

sns.displot(df['age'],kde = True)
```

```python
df["age_mean"]=df['age'].fillna(df['age'].mean())
```

```python
df[["age","age_mean"]]
```

```python
### Mean imputation works well when we have normally distributed data

```

```python
## Median Value imputation -- If we have outliers in the data set we use median

df['age_median']=df['age'].fillna(df['age'].median())
```

```python
df[["age","age_median"]]
```

```python
sns.displot(df['age_median'])
```

```python
## Mode imputation Technique -- > can be used with categorical values


df.head()
```

```python
df['embarked'].isnull().sum()
```

```python
df['embarked'].value_counts()
```

```python
df['embarked'].notna().sum()
```

```python
df['embarked_mode']=df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
df[['embarked','embarked_mode']]
```

```python
### Random sample Imputation
df['age'].isnull().sum()
```

```python
df['age_random']=df['age'].fillna(2)
```

```python
df['age_random'].value_counts()
```
