```python
## It is a technique used to encode categrorical values based on their relationship with the target variable
## This is useful when we have large number of unique categories
```

```python
import pandas as pd
df = pd.DataFrame({
    'city': ['New York', 'London', 'Paris', 'Tokyo', 'New York', 'Paris'],
    'price': [200, 150, 300, 250, 150, 300]
})
```

```python
df
```

```python
df.groupby('city')['price'].mean().to_dict()
```

```python
df['city_encoded']=df['city'].map(df.groupby('city')['price'].mean().to_dict())
```

```python
df['city_encoded']
```

```python
import seaborn as sns
data = sns.load_dataset('tips')
```

```python
data['time_encoded']=data['time'].map(data.groupby('time')['total_bill'].mean().to_dict())
```
