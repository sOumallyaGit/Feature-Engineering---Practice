```python
## Nominal / OHE - is a technique used to represent categorical data as numerical data

## One hot encoding should not be used when a feature has many many categories because it can lead to sparse matrix which can further lead to overfittting

```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

```

```python
## Creating a data frame

df = pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
## Create Instance of one Hot encoder

encoder = OneHotEncoder()
```

```python
## perform fit and transfrom

encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoded_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoded_df],axis=1)
```

```python
import seaborn as sns
data = sns.load_dataset('tips')
```

```python
encoded_sex = encoder.fit_transform(data[['sex']]).toarray()
```

```python
encoded_sex
```

```python
encoded_sex_df = pd.DataFrame(encoded_sex,columns=encoder.get_feature_names_out())
```

```python
encoded_sex_df
```
